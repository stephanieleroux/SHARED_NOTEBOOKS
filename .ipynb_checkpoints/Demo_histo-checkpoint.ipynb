### 1. Read some example monthly data

```python
%matplotlib notebook
%matplotlib inline

import os, user, sys
import matplotlib.pyplot as plt
import pandas as pd
import pickle as pickle
import numpy as np
import seaborn as sns



# Read df (which is a pandas dataframe)
diridat  = "./"
fili     = 'data.M.demo.pkl'
df       = pd.read_pickle(diridat+fili)

# Take the first column as a numpy vector and plot:
datats   = np.array(df["m01"])
```

### 2. Plot (bins on vertical axis)

```python
sns.set_style("darkgrid", {"axes.facecolor": ".96"})

fig1 = plt.figure(figsize=([6,6]),facecolor='white')        

sns.distplot(datats,vertical=True,color="#04b4ae",hist=True)

plt.show()   
fig1.savefig("./testhistoV.pdf", facecolor=fig1.get_facecolor(), edgecolor='none')
```

### 3. Plot (bins on horizontal axis)

```python
sns.set_style("darkgrid", {"axes.facecolor": ".96"})

fig2 = plt.figure(figsize=([6,6]),facecolor='white')        

sns.distplot(datats,vertical=False,color="#04b4ae",hist=True)

plt.show()   
fig2.savefig("./testhistoH.pdf", facecolor=fig2.get_facecolor(), edgecolor='none')
```
